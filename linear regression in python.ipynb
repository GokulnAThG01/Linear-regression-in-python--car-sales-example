```python
# size and year as dependent variables
from statsmodels.tools.tools import add_constant
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as ssn
import io
ssn.set()
multiData = pd.read_csv(io.BytesIO(uploaded['real_estate_price_size_year.csv']));
multiData.describe()

y= multiData['price'];
x= multiData[['size','year']]
x1 = add_constant(x);
result = sm.OLS(y,x1).fit();
result.summary()
```

here the coefficients and the constant have a 'p value' close to zero. hence they are significantly different from zero.


*   higher the F-statistic, higher the significance of our model.
*   in multiple regression, the null hypothesis is that ALL COEFFICIENTS ARE ZERO.

```python
reg = LinearRegression();
# reg is an INSTANCE of the object LinearRegression
reg.fit(xmatrix,y)
```

```python
# to get R squared
reg.score(xmatrix,y)
```

```python
# to get coefficients
reg.coef_
```

```python
# to get the intercept
reg.intercept_
```

```python
# to make prediction
newdata = pd.DataFrame(data= [1740,1760],columns = ['SAT']);
reg.predict(newdata)
```

```python
# plotting the data
plt.scatter(x,y);
yhat = reg.coef_*xmatrix + reg.intercept_;
plt.plot(x,yhat,c ='orange');
plt.xlabel('SAT');
plt.ylabel('GPA');
plt.show()
```

$R^2_{adj.} = 1-(1-R^2)*\frac{n-1}{n-p-1}$

n - number of samples           
p - number of predictors

sklearn doesnt cal R2 with a inbuilt fn !!

```python
import numpy as np
import pandas as pd
import sklearn.linear_model as LinearRegression
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import io
sns.set()
rawdata = pd.read_csv(io.BytesIO(uploaded['1.04.+Real-life+example.csv']))
rawdata.head()
```

## preprocessing
   a lot of information for parameter model can be obtained by a combination of brand, engineV and year....
   
   so the parameter can be dropped

```python
rawdata.describe(include= 'all')
```

```python
data = rawdata.drop(['Model'], axis =1)
# axis = 0 refers to the rows. axis = 1 refers to columns
data.isnull().sum ()    # this is because missing entries are assigned True, rest False
# to find the number of missing values
```

if those with **missing data is less than 5% of the data**, we can simply **remove** such observations

dropna() *function return Index without NA/NaN values. All the missing values are removed and a new object is returned which does not have any NaN values present in it.*

```python
data_no_mv = data.dropna(axis = 0)
data_no_mv.describe()
```

## dealing with outliers


1.   deleting top 1% of the data
2.   log transform...


```python
# the mean price is 19,552 but the max price is 300,000. clearly there are outliers in the dataset
plt.figure(0)
sns.distplot(data_no_mv['Price'])
plt.figure(1)
sns.distplot(data_no_mv['Mileage'])
plt.figure(2)
sns.distplot(data_no_mv['Year'])
plt.figure(3)
sns.distplot(data_no_mv['EngineV'])
```

[how to handle outliers](https://www.sigmamagic.com/blogs/how-to-handle-outliers/)
->check out this article



1.   remove bottom 1% for outliers in year
2.   engine volume greater than 6.5 makes no sense. larger entries are wrong values
3.   remove top 1% for milege and price




```python
# deleting top 1%
q = data_no_mv['Price'].quantile(0.99)
data1 = data_no_mv[data_no_mv['Price']<q]

l = data1['Mileage'].quantile(0.99)
data2 = data1[data1['Mileage']<l]

m = data2['Year'].quantile(0.01)
data3 = data2[data2['Year']>m]

data_cleaned  = data3[data3['EngineV']<6.5].reset_index(drop = True)
data_cleaned.describe(include='all')

# nearly 250 observations are deleted so far! these are problematic ones...
```

## OLS assumptions



1.   linearity using log transformation
2.   checking multicollinearity and removing unnecessary variables -> [When Can You Safely Ignore Multicollinearity?](https://statisticalhorizons.com/multicollinearity)



```python
fig, (a1,a2,a3)= plt.subplots(1,3,sharey=True, figsize = (15,3))
a1.scatter(data_cleaned['Year'],data_cleaned['Price'])
a1.set_title('price vs year')
a2.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
a2.set_title('price vs EngineV')
a3.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
a3.set_title('price vs Mileage')
# dependence of the price on parameters is not linear
```

```python
log_price = np.log(data_cleaned['Price'])
data_cleaned['log price'] = log_price
log_EngineV = np.log(data_cleaned['EngineV'])
data_cleaned['log EngineV']  = log_EngineV

# repeat the plots with log price
plt.figure(0)
fig, (a1,a2,a3)= plt.subplots(1,3,sharey=True, figsize = (15,3))
a1.scatter(data_cleaned['Year'],data_cleaned['log price'])
a1.set_title('log price vs year')
a2.scatter(data_cleaned['EngineV'],data_cleaned['log price'])
a2.set_title('log price vs EngineV')
a3.scatter(data_cleaned['Mileage'],data_cleaned['log price'])
a3.set_title('log price vs Mileage')


```

```python
data_cleaned = data_cleaned.drop(['Price'], axis =1)

```

```python
# it is intuitive that mileage and year have multicollinearity. more the mileage earlier the year......?
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Year','EngineV','Mileage']]
vif = pd.DataFrame()

vif['VIF'] = [variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])] 
vif['features'] = variables.columns
vif
```

```python
data_noMulticollinearity = data_cleaned.drop(['Year'],axis=1)
```

## notes

1) First, once you include model and year, there is multicollinearity. This is one reason why some weights are inflated (like crazy).

2) Second, 'Model' is super segmented. There are 312 models if I remember correctly. Some of them have no more than 5 observations total. This is completely insufficient for any analysis. If there is a single observation in a given model dummy and this observation was an outlier to begin with, the result is rather ugly. What is normally done in such cases is to BUNDLE UP models that are similar. In the end we are hoping to achieve a rather balanced distribution (not too few, not too many observations in each dummy). This operation is called: coarse classing, unfortunately, it is out of the scope of this course (for now).

3) Third, there are outliers that we did not deal with (cause we did not expect to use the  'Model' variable).

Finally, note that there is an easy (but naive) fix for that. If a value is above the maximum in that dataset (say $200,000 or whatever value you find appropriate), you just bind it to the maximum of the dataset. So in your case, you get +infinity, well, just say: this prediction is the maximum of the dataset. In this way, you will get an unrealistically high prediction, which however is still valid (and not +infinity).
